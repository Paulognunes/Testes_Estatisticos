## Intervalo de Confiança
---

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Encontrar a nota média dos filmes do TMDB é uma tarefa relativamente simples, basta somar todas as notas e dividir pela quantidade. Porém, se quisermos saber qual é a nota dos filmes em geral estando, ou não, dentro do dataset ? <p>
    
<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Isso pode ser feito utilizando os testes estatísticos. Os dois principais são o ttest e o ztest. O ttest é mais recomendado quando o conjunto de dados é relativamente pequeno. Já o ztest, é amplamente usado em conjunto de dados maiores. Define-se o intervalo de confiança desejado e passa-se como parâmetro o dataset. Como retorno, a função retornará a média dos filmes em geral, supondo que todos os filmes apresentem semelhanças com o dataset utilizado.<p>

```python
# Intervalo de confiança das notas dos filmes utilizando o ztest
zconfint(tmdb['vote_average'])
```

```python
# Intervalo de confiança das notas dos filmes utilizando o ttest
desc_nota_media = DescrStatsW(tmdb['vote_average'])
desc_nota_media.tconfint_mean()
```

<p style='line-height: 2; text-align: justify; text-indent: 35px;'>É possivel também verificar a validade de uma hipótese. Por exemplo, queremos verificar se a média das notas dos filmes no mundo é igual à 6,0. Para isso, basta aplicar um ztest e informar o valor que achamos que é o correto. Como retorno, a função retorna um pvalue. Se esse valor for menor que o intervalo de confiança utilizado, em geral 5%, então descarta-se a hipótese. <p>

```python
# Teste estatístico ztest
media_geral = ztest(tmdb['vote_average'], value=5)
media_geral
```

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Também podemos gerar um intervalo de confiança entre as notas de um filme específico e todos os demais do dataset por meio do ztest. Abaixo, iremos comparar o intervalo de confiança entre os filmes lançados no ano de 2015 em relação à todos os filmes do dataset. <p>

```python
# Transformando a coluna de release em um objeto do datetime
tmdb['release_date'] = pd.to_datetime(tmdb['release_date'], format='%Y-%m-%d')
```

```python
# Selecionando os filmes de 2015
filmes_2015 = tmdb.loc[(tmdb['release_date'] >= '2015-01-01') & (tmdb['release_date'] < '2015-12-31')] 
filmes_2015
```

```python
# Intervalo de confiança entre duas amostras 
zconfint(filmes_2015['vote_average'], tmdb['vote_average'])
```

```python
# Pvalue
ztest(filmes_2015['vote_average'], tmdb['vote_average'])
```

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Dado os valores do intervalo de confiança entre os filmes lançados em 2015 com os demais do dataset, conclui-se que a sua média da distribuição dos votos é menor que a média de distribuição dos votos global do dataset.<p>

```python
# Boxplot
plt.figure(figsize=(12,8))
plt.boxplot([tmdb['vote_average'], filmes_2015['vote_average']], labels=['Todos os Filmes do Dataset', 'Filmes lançados em 2015'])
plt.title("Distribuição das notas", fontsize=20)
plt.ylabel("Nota dos filmes", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

## Testes Não-Paramétricos
---

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Tanto o ztest quanto o ttest só podem ser aplicados corretamente se a distribuição dos elementos aproxima-se de uma distribuição normal. Para isso, podemos utilizar a função normaltest do scipy para testar se a distribuição dos elementos é, ou não, normal. Assim, vamos verificar se a distribuição das notas dos filmes lançados em 2015 é normal. Se o valor do pvalue for menor que o intervalo de confiança empregado, normalmente 95%, então considera-se uma distribuição normal se o pvalue for maior ou igual à 0,05. <p>

```python
_, pvalue = normaltest(filmes_2015['vote_average'])
pvalue
```

<p style='line-height: 2; text-align: justify; text-indent: 35px;'>  <p>

```python
# Histograma do vote_average dos filmes
ax = sns.distplot(filmes_2015["vote_average"], norm_hist=False)
ax.figure.set_size_inches(20, 6)
ax.set_title('Média de Notas dos Filmes (2015)', fontsize=20)
ax.set_xlabel('Nota Média', fontsize=16)
ax.set_ylabel('Número de Votos', fontsize=16)
ax
```

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Ao realizarmos o teste de normalidade no dataset todo, constata-se que ele não apresenta uma distribuição normal pois seu pvalue é menor que 0,05. Logo, não é posssível realizar os ttest e o ztest. Nesse conjunto de dados, é necessário aplicar uma categoria diferente de teste, que é denominado de testes não-paramétricos. <p>

```python
_, pvalue = normaltest(tmdb['vote_average'])
pvalue
```

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Um dos testes paramétricos mais utilizados é o teste de Wilcoxon (ranksums). Se o valor do pvalue &gt; 0,05 conclui-se que ambos os conjuntos vieram da mesma distribuição. Se o pvalue &le; e 0,05 conclui-se que os valores de um conjunto apresentam uma tendência de ser maiores que os do outro conjunto. <p>

```python
_, pvalue = ranksums(filmes_2015['vote_average'], tmdb['vote_average'])
pvalue
```
